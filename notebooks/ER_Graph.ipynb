# Generating ER Graphs

```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
def generate_graph(n, p, seed = None, directed = None):
    G= nx.erdos_renyi_graph(n, p, seed, directed)
    return G
```

```python
# Function to generate and draw graph
def generate_and_draw_graph(n, p, seed, ax, title):
    G = nx.erdos_renyi_graph(n=n, p=p, seed=seed)  # Generates an Erdos-Renyi graph
    nx.draw(G, with_labels=True, ax=ax, node_color='skyblue', font_size=10, font_weight='bold')
    ax.set_title(title, fontsize=14)
```

```python
# Create a figure with subplots (1 row, 3 columns)
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

# Generate and plot the three graphs
generate_and_draw_graph(n=10, p=0.5, seed=191, ax=axes[0], title="Graph 1: Nodes = 10")
generate_and_draw_graph(n=20, p=0.5, seed=42, ax=axes[1], title="Graph 2: Nodes = 20")
generate_and_draw_graph(n=50, p=0.5, seed=123, ax=axes[2], title="Graph 3: Nodes = 50")

# Adjust layout for better spacing
plt.tight_layout()

# Show the final plot
plt.show()
```
