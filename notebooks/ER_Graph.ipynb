# Generating ER Graphs

```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
def generate_graph(n, p, seed = None, directed = None):
    G= nx.erdos_renyi_graph(n, p, seed, directed)
    return G
```

```python
G = generate_graph(n= 10, p=0.5)

nx.draw(G, with_labels=True)
plt.show()
```
